```python


import torch
from sbi.inference import NPE

# define shifted Gaussian simulator.
def simulator(θ): return θ + torch.randn_like(θ)
# draw parameters from Gaussian prior.
θ = torch.randn(1000, 2)
# simulate `data`
x = simulator(θ)

# choose sbi method and train
inference = NPE()
inference.append_simulations(θ, x).train()

# do inference given observed data
x_o = torch.ones(2)
posterior = inference.build_posterior()
samples = posterior.sample((1000,), x=x_o)
print(samples)

```

```python
from sbi.inference import NPE
import torch
from sbi.utils import BoxUniform

# ----- 1. PRIOR -----
prior = BoxUniform(low=torch.tensor([-1.0]), high=torch.tensor([1.0]))

# ----- 2. SIMULATOR -----
def simulate(theta):
    return theta**2 + 0.1 * torch.randn_like(theta)

# ----- 3. Simulations -----
num_simulations = 1000
theta = prior.sample((num_simulations,))
x = simulate(theta)

# ----- 4. NPE -----
inference = NPE(prior)

# Train the density estimator
density_estimator = inference.append_simulations(theta, x).train()

# Build the posterior from the trained density estimator
posterior = inference.build_posterior(density_estimator)

```
